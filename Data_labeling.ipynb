```python
import pandas as pd
import re
```

```python
df = pd.read_excel('unigram_freq_ATannotations.xlsx')
df.head()
```

```python
date_reg_exp = re.compile('(\d+[-/]\d+[-/]\d+)')
year = re.compile('([1,2]{1}[0,9]{1}\d{2}$)')
months = ['january', 'febuary', 'march', 'april','june', 'july', 'august', 'september', 'october', 'november', 'december']
my_list = []

for i in range(len(df['Word'])):
    if 'datetime.datetime' in str(type(df['Word'][i])):
        df.loc[i, 'Category'] = 'Date'
        continue
    if date_reg_exp.match(str(df['Word'][i])):
        df.loc[i, 'Category'] = 'Date'
        continue
    if year.match(str(df['Word'][i])):
        if re.match('\d{4}$', str(df['Word'][i])):
            if int(df['Word'][i])>2060 or int(df['Word'][i])<1940:
                df.loc[i, 'Category'] = 'Amount'
                continue
            else:
                df.loc[i, 'Category'] = 'Date' 
                continue
    if any(x in str(df['Word'][i]).lower() for x in months):
        df.loc[i, 'Category'] = 'Date'
```

```python
amount_reg_exp = re.compile('\d{1,5}$')
amounts = ['million', 'billion']

for i in range(len(df['Word'])):
    if amount_reg_exp.match(str(df['Word'][i])) and str(df['Category'][i]) == 'nan':
        df.loc[i, 'Category'] = 'Amount'
        continue
    if any(x in str(df['Word'][i]).lower() for x in amounts):
        df.loc[i, 'Category'] = 'Amount'
        
```

```python
degree_reg_exp = re.compile('\w+ly$')

for i in range(len(df['Word'])):
    if degree_reg_exp.match(str(df['Word'][i])):
        df.loc[i, 'Category'] = 'Degree'
```

```python
import nltk
word_reg_exp = re.compile('^\w{4,8}$')
word_list = []
for i in range(len(df['Word'])):
    if word_reg_exp.match(str(df['Word'][i])):
        word = str(df['Word'][i])
        word_list.append(word)

tag_list = nltk.pos_tag(word_list)
```

```python
verb_list = []
for word, tag in tag_list:
    if 'VB' in tag:
        verb_list.append(word)
```

```python
for i in range(len(df['Word'])):
    if word_reg_exp.match(str(df['Word'][i])):
        if str(df['Word'][i]) in verb_list:
            df.loc[i, 'Category'] = 'Verb'
```
